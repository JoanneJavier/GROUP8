```python
#import dependencies
import pandas as pd 
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
ca_weekly = pd.read_csv("Resources/weekly_gas_prices_ca_2.csv")
ca_weekly = ca_weekly.set_index("Date")
ca_weekly = ca_weekly.loc["2017-09-04":"2018-09-03",:]
```

```python
#drop nan values in the df
ca_weekly = ca_weekly.dropna()
```

```python
#plot graph for weekly gas price in the whole state of CA 

#data = ca_weekly["Price"]
#weeks = ca_weekly["Date"]
#week_data = weeks.dropna()
#week_list = week_data.tolist()
#tick_locations = [value for value in week_list]
#plt.xticks(tick_locations, week_list)
ca_weekly["Price"] = ca_weekly["Price"] * 10

plt.figure(figsize=(20,10))
plt.plot(ca_weekly)
plt.title("Weekly Gas Prices in California from 2013-2018")
plt.ylabel("Gas Price (Dollars per Gallon)")
plt.xticks(rotation=90)
plt.grid()

plt.show()

#clean up xticks 
```
