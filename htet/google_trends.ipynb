```python
from pytrends.request import TrendReq
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import json
from config import gkey
```

```python
pytrends = TrendReq(hl='en-US', tz=360)
```

```python
kw_list = ['Tesla model']
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='US-CA',gprop='')
```

```python
tesla_by_region = pytrends.interest_by_region(resolution = 'CITY')
```

```python
tesla_over_time = pytrends.interest_over_time()
tesla_over_time.head()
```

```python
plt.plot(tesla_over_time["Tesla model"])
```

```python
plt.figure(figsize=(20,10))
ecar, = plt.plot(ecar_over_time["Electric car"])
tesla, = plt.plot(tesla_over_time["Tesla model"])
plt.legend((ecar,tesla),("Electric Car","Tesla"), loc="best")
plt.title("Google Search Trends for Electric Car Related Terms from 2013 to 2018")
plt.xlabel("Year")
#clarify the scale for google trend
plt.ylabel("Interest")
plt.grid()

plt.savefig("Images/trend_over_time.png")
```
