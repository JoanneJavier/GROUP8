```python
from pytrends.request import TrendReq
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import json
from config import gkey
```

```python
pytrends = TrendReq(hl='en-US', tz=360)

kw_list = ['Tesla', 'Electric car', 'Prius', 'Porsche Electric Car']
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='US-CA',gprop='')
```

```python
interest_by_region = pytrends.interest_by_region(resolution = 'CITY')
interest_by_region.head()
```

```python
interest_over_time = pytrends.interest_over_time()
interest_over_time.head()
```

```python
plt.figure(figsize=(20,10))
plt.plot(interest_over_time)
```
